### Multiplying 3 to a list vs array

```python
d = [11, 12, 13, 14, 15]
type(d)
```

```python
d * 3
```

```python
# In list need to apply for loop to calculations
e = []
for i in d:
    e.append(i * 3)
```

```python
e
```

```python
# Multiplying by 3 using arrays
d1 = np.array(d)
d1
```

```python
type(d1)
```

```python
# Brodcasting in numpy arrays
e1 = d1 * 3
```

```python
e1
```

```python
d1 + 5
```

```python
4*d1 + 5
```

```python
d1**2
```

```python
s1 = d1**(1/2)
s1
```

```python
s1.dtype
```

```python
d1.dtype
```

### Products of element inside list vs numpy array

```python
g = [51, 101, 48, 49, 50, 53]
type(g)
```

```python
# For list need to apply for loop
p = 1
for i in g:
    p = p * i
```

```python
p
```

```python
sum(g)
```

```python
# Numpy array use np.prod
p1 = np.prod(g)
p1
```
