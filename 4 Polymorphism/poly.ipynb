# How does a triangle form
1. Triangle Inequality Sum of any two sides should be greater than third
2. Perimeter is sum of all sides
3. Area is calculated by Herons formula



```python
from dataclasses import dataclass
```

```python
@dataclass
class Triangle:
    a: int | float
    b: int | float
    c: int | float

    def is_triangle(self):
        if ((self.a + self.b > self.c) and 
            (self.b + self.c > self.a) and 
            (self.c + self.a > self.c)):
            return True
        else:
            return False
        
    def perimeter(self):
        if self.is_triangle():
            p = self.a + self.b + self.c
            return p
        else:
            raise ValueError(f"Given sides {self.a}, {self.b}, {self.c} do not form a triangle")
    
    def area(self):
        s = self.perimeter() / 2
        a = (s * (s - self.a) * (s - self.b) * (s - self.c))**(1 / 2)
        return a
```

### Polymorphism in user defined classes
Method Overriding

```python
class India:

    def capital(self):
        print("New Delhi is capital of India")

    def language(self):
        print("Hindi is widely spoken in India")
```

```python
class USA:

    def capital(self):
        print("Washington D.C. is capital of USA")

    def language(self):
        print("English is widely spoken in USA")
```

```python
class France:

    def capital(self):
        print("Paris is capital of France")

    def language(self):
        print("French is widely spoken in France")
```

```python
c1 = India()
type(c1)
```

### Calulate area and perimeter for different shapes

```python
@dataclass
class Rectangle:
    height: int | float
    width: int | float

    def perimeter(self):
        return 2 * (self.height + self.width)
    
    def area(self):
        return self.width * self.height
```

```python
from math import pi
```

```python
pi
```

```python
@dataclass
class Circle:
    radius: int | float

    def perimeter(self):
        return 2 * pi * self.radius
    
    def area(self):
        return pi * (self.radius ** 2)
```

```python
shapes = [
    Triangle(3, 4, 5),
    Circle(7),
    Rectangle(10, 20),
    Circle(20),
    Triangle(12, 13, 14),
    Rectangle(15, 12)
]
```

```python
shapes
```

```python
for i in shapes:
    print(i, type(i))
    print(f"Perimeter : {i.perimeter():.2f}")
    print(f"Area : {i.area():.2f}")
    print("\n", "="*50, "\n")
```
